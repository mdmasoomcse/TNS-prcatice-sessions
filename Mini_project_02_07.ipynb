```python
df['Effiecency']=(df['Range']/df['Battery_capacity'])*100
df
```

```python
#to classify vehicle on efficiency
def classify_efficiency(efficiency_value):
  if efficiency_value > 8:
    return "Excellent"
  elif efficiency_value >= 6 and efficiency_value < 8:
    return "Good"
  elif efficiency_value > 4 and efficiency_value <= 6:
    return "Average"
  else:
    return "Poor"

df['Efficiency_Class'] = df['Effiecency'].apply(classify_efficiency)
df
```
