```python
import pandas as pd

data = {
    "Name": ["Rahul", "Sneha", "Amit", "Priya", "Kiran", "Divya"],
    "Department": ["Sales", "IT", "IT", "HR", "Sales", "Finance"],
    "City": ["Hyderabad", "Bangalore", "Hyderabad", "Chennai", "Bangalore", "Bangalore"],
    "Satisfaction": ["Low", "High", "Medium", "High", "Low", "Medium"]
}

df = pd.DataFrame(data)
df
```

Label encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df["Department_Label"] = le.fit_transform(df["Department"])
df["City_Label"] = le.fit_transform(df["City"])
df
```

Ordinal Encoding

```python
order = {"Low": 1, "Medium": 2, "High": 3}

df["Satisfaction_Ordinal"] = df["Satisfaction"].map(order)
df
```

Frequency Encoding

```python
freq = df["City"].value_counts()

df["City_Freq"] = df["City"].map(freq)
df
```

Target Encoding

```python
# Create a dummy target variable
df["Target"] = [1, 0, 1, 0, 1, 0]

# Calculate mean target per category
target_mean = df.groupby("Department")["Target"].mean()

df["Department_Target_Encoded"] = df["Department"].map(target_mean)
df
```
