```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded=files.upload()
```

```python
df = pd.read_csv('Global YouTube Statistics.csv', encoding='latin1')


plt.show()
```

```python
a=df.sort_values(by='video views',ascending=False).head(5)

plt.bar(a['Youtuber'],a['video views'])
plt.xlabel('Youtuber')
plt.ylabel('video views')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.title('Top 10 Youtubers by video views')
plt.figure(figsize=(8,10))
plt.show()
```

```python
top_subscribers = df.sort_values(by='subscribers', ascending=False).head(10)
sns.barplot(x='Youtuber', y='subscribers', data=top_subscribers)
plt.title('Top 10 Youtubers by Subscribers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
top_views=df[['Youtuber','video views']].sort_values(by='video views',ascending=False).head(10)
plt.pie(top_views['video views'],labels=top_views['Youtuber'],autopct='%1.1f%%')
plt.title('Top 10 Youtubers by video views')
plt.show()
```
