```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

flights=sns.load_dataset("flights")
flights.head()
```

```python
sns.scatterplot(x="month",y="passengers",data=flights)
plt.legend()
plt.figure(figsize=(10,5))
```

```python
import seaborn as sns
iris=sns.load_dataset("iris")
iris.head(15)
```

```python
sns.barplot(data=iris)
```

```python
penguins=sns.load_dataset("penguins")
penguins.fillna(penguins.mean(numeric_only=True), inplace=True)
penguins["sex"].fillna("Male",inplace=True)
penguins.head()
```

```python
sns.regplot(x="flipper_length_mm",y="body_mass_g",data=penguins)
```
