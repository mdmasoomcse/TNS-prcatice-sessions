```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Presentation_Marks_Report.xlsx')
clean_df = df.dropna().copy()
clean_df[["Technical", "Communication", "Attire"]] = clean_df[["Technical", "Communication", "Attire"]].astype(int)

# Total Marks = Technical + Communication + Attire
clean_df["Total Marks"] = clean_df["Technical"] + clean_df["Communication"] + clean_df["Attire"]

print(f"Removed {len(df) - len(clean_df)} records with missing values.")
print(f"Clean records: {len(clean_df)}")
clean_df.reset_index(drop=True, inplace=True)
clean_df
```

```python
clean_df.describe()[["Technical", "Communication", "Attire", "Total Marks"]]
```

```python
def top5(column):
    return (clean_df.sort_values([column, "Total Marks", "UID"], ascending=[False, False, True])
                     .head(5)[["UID", "Participant Name", column]]
                     .reset_index(drop=True))

top5_total = top5("Total Marks")
top5_tech = top5("Technical")
top5_comm = top5("Communication")
top5_attire = top5("Attire")

print("Top 5 - Total Marks"); display(top5_total)
print("Top 5 - Technical"); display(top5_tech)
print("Top 5 - Communication"); display(top5_comm)
print("Top 5 - Attire"); display(top5_attire)
```

```python
plt.figure(figsize=(8, 5))
ax = sns.barplot(data=top5_total, x="Total Marks", y="Participant Name", palette="crest")
ax.set_title("Top 5 Students — Total Marks", fontsize=13, fontweight="bold")
ax.set_xlabel("Total Marks")
ax.set_ylabel("")
for i, v in enumerate(top5_total["Total Marks"]):
    ax.text(v + 0.1, i, str(v), va="center", fontweight="bold")
plt.tight_layout()
plt.show()
```

### 4.2 Top 5 — Technical, Communication, Attire (side by side)

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
data_map = [("Technical", top5_tech, "flare"), ("Communication", top5_comm, "mako"), ("Attire", top5_attire, "crest")]

for ax, (col, data, palette) in zip(axes, data_map):
    sns.barplot(data=data, x=col, y="Participant Name", palette=palette, ax=ax)
    ax.set_title(f"Top 5 — {col}", fontweight="bold")
    ax.set_xlabel(col)
    ax.set_ylabel("")
    for i, v in enumerate(data[col]):
        ax.text(v + 0.05, i, str(v), va="center", fontweight="bold", fontsize=9)

plt.tight_layout()
plt.show()
```

### 4.3 Score Distributions

```python
fig, axes = plt.subplots(1, 4, figsize=(18, 4))
cols = ["Technical", "Communication", "Attire", "Total Marks"]
colors = ["#4C72B0", "#55A868", "#C44E52", "#8172B2"]

for ax, col, color in zip(axes, cols, colors):
    sns.histplot(clean_df[col], kde=True, ax=ax, color=color, bins=10)
    ax.set_title(col, fontweight="bold")

plt.tight_layout()
plt.show()
```
