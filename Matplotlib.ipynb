```python
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd

data = {
    "Month": ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
    "Sales": [1200,1350,1500,1600,1750,1900,2100,2050,2200,2350,2500,2800],
    "Expenses": [800,850,900,950,1000,1100,1200,1180,1250,1300,1400,1500],
    "Customers": [150,170,180,200,220,240,260,255,280,300,320,350],
    "Rating": [4.1,4.3,4.4,4.5,4.6,4.7,4.8,4.7,4.9,4.8,4.9,5.0]
}

df = pd.DataFrame(data)
df.to_csv("sales_data.csv", index=False)
```

```python
a=df['Rating']
b=df['Expenses']
c=df['Customers']
d=df['Sales']
x=df['Month']
y=df["Sales"]

plt.plot(x,y)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Sales Analysis")
plt.grid(True)
plt.legend("Sales")
plt.show()
```

```python
plt.bar(c,b)
plt.xlabel("Coustomers")
plt.ylabel("Expenses")
plt.show()
```

```python
plt.scatter(x,y)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
plt.hist(y)
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```
