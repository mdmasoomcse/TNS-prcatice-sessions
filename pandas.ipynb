```python
import pandas as pd
data={
    "animal":["dog","cat","horse","camel","snake","cat"],
    "type":["mammal","mammal","mammal","mammel","reptile","Cat"],
    "numbers":[1,2,3,4,5,6]
}
data1={
    "animal":["dog","cat","horse","camel","snake","cat"],
    "type":["mammal","reptile","mammal","mammel","reptile","fox"],
    "numbers":[1,2,3,4,5,6]
}
fr=pd.DataFrame(data)
print(fr)

fr1=pd.DataFrame(data1)
print(fr1)

merged_fr=pd.merge(fr,fr1,on=["animal","type"])
print(merged_fr)

cocat=pd.concat([fr,fr1])
print(cocat)
```

```python
hospital={
    "Patient_id":[501,502,503,507,504,509],
    "Name":["rajesh","Ravi","koushik","nandu","satwik","sai"],
    "Age":[14,52,14,32,22, None], # Added None to make its length 6
    "Diesease":["typhoid","malaria","dengue","mental_illness","dipression","HIV"],
    "Days-Admitted":[5,40,7,7,9,10],
}
data_frame=pd.DataFrame(hospital)
print(data_frame)

duplicate_mask=data_frame['Days-Admitted'].duplicated()
print("Duplicates in 'Days-Admitted' column:", duplicate_mask)

add_dataframe=pd.DataFrame({"cost_per_day in thousand":[5,12,55,75,55,12]})
print(add_dataframe)

merged_df=pd.concat([data_frame,add_dataframe],axis=1)
print(merged_df)

# mean=data_frame["add_dataframe"].mean()
# print("your mean",mean)

sum_cost=merged_df["cost_per_day in thousand"].sum()
print("your sum is:",sum_cost)

avg_cost=merged_df["cost_per_day in thousand"].mean()
print(avg_cost)

max=merged_df["cost_per_day in thousand"].max()
print(max)

min=merged_df["cost_per_day in thousand"].min()
print(min)
```
